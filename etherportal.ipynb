```python
#@title 2. Download Model
Model_to_download="Anythingv3" #@param [ "Anythingv3", "Eimis", "EvtV3", "ACertain", "ElysiumV3", "Wlop-any", "WaifuDiffusion-Vae", "LeafMix", "Abyss7thLayer", "Healys", "LigneClaire", "KriboMix", "InizioSkinjob", "PantyStocking", "WaifuBodyBlenderMix", "StuffyMix", "Berrymix", "Systemy-CSR", "ElyOrangemix", "AnyTwamMix", "DarkberryMix", "Grapefruit", "Mitsudoue"]
#@markdown Confused on what model to choose? Look for examples and comparison [here](https://github.com/etherealxx/etherportal-webui-colab/blob/master/ModelComparison.md).
#@markdown <br/>If the download speed is slow (Below 10MiB/s), press the stop button then re-run the cell)

import os
from IPython.utils import capture
import requests
from tqdm import tqdm

folder = '/content/stable-diffusion-webui/models/Stable-diffusion/'

def checkdownloaded(pathx):
  if os.path.exists(pathx):
    if pathx.endswith("tmp"):
      os.remove(pathx)
    else:
      print(f'[1;32mThat model is already downloaded.')
      print('[0m')
      return 2

def download_hf(filename, link):
  hfpath = os.path.join(folder,filename)
  check = checkdownloaded(hfpath)
  if check == 2: 
    if filename=='Anything-V3.0.ckpt' and not os.path.exists(folder+'Anything-V3.0.vae.pt'): 
      !gdown https://huggingface.co/Linaqruf/anything-v3.0/resolve/main/Anything-V3.0.vae.pt -O /content/stable-diffusion-webui/models/Stable-diffusion/Anything-V3.0.vae.pt
    return

  print(f'\033[1;32mDownloading {Model_to_download} Model...\033[0m')
  !gdown {link} -O /content/stable-diffusion-webui/models/Stable-diffusion/{filename}
  if filename=='Anything-V3.0.ckpt':
    !gdown https://huggingface.co/Linaqruf/anything-v3.0/resolve/main/Anything-V3.0.vae.pt -O /content/stable-diffusion-webui/models/Stable-diffusion/Anything-V3.0.vae.pt
  print(f'\033[1;32mModel Downloaded.\033[0m')

def download_civit(fname, url):
    print(f'\033[1;32mDownloading {Model_to_download} Model...\033[0m')
    civitpath = os.path.join(folder,fname)
    check = checkdownloaded(civitpath)
    if check == 2: return

    head_response = requests.head(url)
    total = int(head_response.headers.get('Content-Length', 0))
    if os.path.exists(civitpath):
        downloaded_size = os.path.getsize(civitpath)
        if total != 0 and downloaded_size != total:
            os.remove(civitpath)
            print(f'Removed incomplete download: {fname}')
    else:
        downloaded_size = 0

    if not os.path.exists(civitpath):
        pathtemp = civitpath + 'bdgh'
        filenamebefore = os.path.basename(civitpath)
        filenameafter = os.path.basename(pathtemp)
        name_parts = filenameafter.rsplit('.', 1)
        ext = name_parts[1].rstrip('bdgh')
        combined = name_parts[0] + '.' + ext
        print(f'The model will be saved as {filenamebefore}')
        with open(pathtemp, 'wb') as file, tqdm(
            desc=pathtemp,
            total=total,
            unit='iB',
            unit_scale=True,
            unit_divisor=1024,
            initial=downloaded_size,
        ) as bar:
            resp = requests.get(url, stream=True)
            for data in resp.iter_content(chunk_size=1024):
                size = file.write(data)
                bar.update(size)
        # Check if the file has been downloaded successfully
        downloaded_size = os.path.getsize(pathtemp)
        if total != 0 and downloaded_size != total:
            # Remove the file if it has not been downloaded successfully
            os.remove(pathtemp)
            print(f'Removed incomplete download: {fname}')
            return
        %cd {folder}
        os.rename(filenameafter, combined)
        print(f'\033[1;32mModel Downloaded.\033[0m')


print('[1;32mInstalling webui...')
with capture.capture_output() as cap:
  %cd /content
  !if ! [ -d stable-diffusion-webui ]; then git clone -b v1.6 https://github.com/camenduru/stable-diffusion-webui; fi
print('[0m')
!if [ -f {folder}'Put Stable Diffusion checkpoints here.txt' ]; then rm {folder}'Put Stable Diffusion checkpoints here.txt'; fi

%cd {folder}
    #sfw
if Model_to_download=='Anythingv3':
  download_hf('Anything-V3.0.ckpt', 'https://huggingface.co/Linaqruf/anything-v3.0/resolve/main/Anything-V3.0-pruned-fp32.ckpt')
if Model_to_download=='WaifuDiffusion-Vae':
  download_hf('WD-vae.vae.pt', 'https://huggingface.co/hakurei/waifu-diffusion-v1-4/resolve/main/vae/kl-f8-anime2.ckpt')
if Model_to_download=='Eimis':
  download_hf('EimisAnimeDiffusion.ckpt', 'https://huggingface.co/eimiss/EimisAnimeDiffusion_1.0v/resolve/main/EimisAnimeDiffusion_1-0v.ckpt')
if Model_to_download=='EvtV3':
  download_hf('Evt_V3.ckpt', 'https://huggingface.co/haor/Evt_V3/resolve/main/Evt_V3_ema.ckpt')
if Model_to_download=='ACertain':
  download_hf('ACertainModel.ckpt', 'https://huggingface.co/JosephusCheung/ACertainModel/resolve/main/ACertainModel.ckpt')
if Model_to_download=='ElysiumV3':
  download_hf('Elysium_Anime_V3.safetensors', 'https://huggingface.co/hesw23168/SD-Elysium-Model/resolve/main/Elysium_Anime_V3.safetensors')
if Model_to_download=='Wlop-any':
  download_hf('wlop-any.ckpt', 'https://huggingface.co/SirVeggie/wlop/resolve/main/wlop-any.ckpt')
if Model_to_download=='LeafMix':
  download_hf('Leaf-pruned.safetensors', 'https://huggingface.co/etherealxx/anyleaf-temp/resolve/main/Leaf-pruned.safetensors')
if Model_to_download=='Abyss7thLayer':
  download_hf('Abyss_7th_anime_v1.1.ckpt', 'https://huggingface.co/syaimu/7th_Layer/resolve/main/Abyss_7th_anime_v1.1.ckpt')
if Model_to_download=='Healys':
  download_civit('healysAnimeBlend_17_2.ckpt', 'https://civitai.com/api/download/models/1498')
if Model_to_download=='LigneClaire':
  download_hf('ligne_claire_anime_diffusion_v1.safetensors', 'https://huggingface.co/breakcore2/ligne_claire_anime_diffusion/resolve/main/ligne_claire_anime_diffusion_v1.safetensors')
if Model_to_download=='KriboMix':
  download_civit('kribomixv1Float16.ckpt', 'https://civitai.com/api/download/models/1289')
if Model_to_download=='InizioSkinjob':
  download_hf('inizio_skinjob.safetensors', 'https://huggingface.co/Cinnamomo/inizio/resolve/main/safetensor/inizio_skinjob.safetensors')
if Model_to_download=='PantyStocking':
  download_civit('digimDsPantyStocking.safetensors', 'https://civitai.com/api/download/models/1560')
if Model_to_download=='WaifuBodyBlenderMix':
  download_civit('waifubodyblenderMix_v1.safetensors', 'https://civitai.com/api/download/models/1838')
if Model_to_download=='StuffyMix':
  download_civit('stuffymix.safetensors', 'https://civitai.com/api/download/models/3376')

    #nsfw
if Model_to_download=='BerryMix':
  download_hf('berry_mix.ckpt', 'https://huggingface.co/misobarisic/berry-mix/resolve/main/berry_mix.ckpt')
if Model_to_download=='Systemy-CSR':
  download_hf('CSRmodel.ckpt', 'https://huggingface.co/etherealxx/systemy-csrmodel-cutesexyrobutts/resolve/main/CSRmodel_1-CSRmodel_0-Weighted_sum-merged.ckpt')
if Model_to_download=='ElyOrangemix':
  download_hf('ElyOrangeMix.ckpt', 'https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/ElyOrangeMix.ckpt')
if Model_to_download=='AnyTwamMix':
  download_civit('anytwam11.safetensors', 'https://civitai.com/api/download/models/4527')
if Model_to_download=='DarkberryMix':
  download_civit('darkberryMix.ckpt', 'https://civitai.com/api/download/models/1992')
if Model_to_download=='Grapefruit':
  download_civit('grapefruitv22.safetensors', 'https://civitai.com/api/download/models/4644')
if Model_to_download=='Mitsudoue':
  download_civit('mitsudoue7k.safetensors', 'https://civitai.com/api/download/models/1454')

files = os.listdir(folder)
for file in files:
    if file.endswith('tmp') or file.endswith('bdgh'):
        file_path = os.path.join(folder, file)
        print(f"Removed incomplete download: {file_path}")
        os.remove(file_path)

print('[1;32mList of installed models:')
print('[0m')
!ls {folder}
```

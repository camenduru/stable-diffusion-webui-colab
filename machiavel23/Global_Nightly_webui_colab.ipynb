```python
################
##### FORM #####
################

#@title #Customize your settings then run the cell
#@markdown ##*Choose a model*
#@markdown <details><summary>Click here to show/hide models' specificities</summary><table><thead><tr><th>Model</th><th>All in One Pixel Model</th><th>Analog Diffusion</th><th>Arcane Diffusion 3</th><th>Cinematic Diffusion</th><th>Comic Diffusion</th><th>Comics Blend</th><th>Complex Lineart</th><th>Cyberpunk Anime Diffusion</th><th>Dreamlike Diffusion</th><th>Food Crit</th><th>Illuminati v1.1</th><th>Illustration Diffusion</th><th>Inkpunk</th><th>Knollingcase</th><th>Midjourney v4</th><th>Mo Di Diffusion</th><th>Papercut</th><th>Popup Book</th><th>Riffusion</th><th>SamDoesArt</th><th>Stable Diffusion 2 Depth</th><th>Stable Diffusion 2.1</th><th>Synthwave punk v2</th><th>Van Gogh Diffusion</th><th>Vintedois Diffusion</th><th>Woolitize</th></tr></thead><tbody><tr><td>Special notes</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;pixelsprite<br>&nbsp;&nbsp;&nbsp;&nbsp;16bitscene</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;analog style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;arcane style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;syberart</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;harliebo artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;holliemengert artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;marioalberti artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;pepelarraz artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;andreasrocha artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;jamesdaly artstyle</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;ComplexLA style<br>&nbsp;&nbsp;&nbsp;&nbsp;nvinkpunk<br>&nbsp;&nbsp;&nbsp;&nbsp;marioalberti artstyle<br>&nbsp;&nbsp;&nbsp;&nbsp;ghibli style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;ComplexLA style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;dgs illustration style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;dreamlikeart style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;food_crit</td><td>Negative prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;nrealfixer<br>&nbsp;&nbsp;&nbsp;&nbsp;nartfixer<br>&nbsp;&nbsp;&nbsp;&nbsp;nfixer</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;holliemengert artstyle</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;nvinkpunk style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;knollingcase</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;mdjrny-v4 style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;modern disney style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;PaperCut</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;popupBook</td><td>Use with the script :<br>&nbsp;&nbsp;&nbsp;&nbsp;Riffusion Audio Generator</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;SamDoesArt</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;FloralMarble-400</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;dblx<br>&nbsp;&nbsp;&nbsp;&nbsp;AnimeScreenCap</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;snthwve style<br>&nbsp;&nbsp;&nbsp;&nbsp;nvinkpunk</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;lvngvncnt style</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;estilovintedois</td><td>Positive prompt:<br>&nbsp;&nbsp;&nbsp;&nbsp;woolitize style</td></tr></tbody></table></details>
#@markdown <details><summary>Click here to show/hide which models download their own VAE, along with comparison grids.</summary><table><thead><tr><th>VAE</th><th>Models that will automatically download it</th></tr></thead><tbody><tr><td>Anything v3</td><td>Abyss Orange Mix 2, Anything v3, Counterfeit 2.5, Monstermash 4/6 + Any v3, Open Anime Journey, Orange Cocoa Mix 2 HARD, Yohan Diffusion</td></tr><tr><td>Anything v4</td><td>Anything v4.5</td></tr><tr><td>Grapefruit</td><td>Grapefruit</td></tr><tr><td>kl_f8_anime2</td><td>Color Bomb Mix, Defmix, Dpepmkp, Elysium Anime 2, Face Bomb Mix, Faetastic v1, Hyper Bomb Mix, Magical Mix 1/2, Pastel Mix (fp16/32), PVC, Waifu Diffusion (all), Yiffy Mix</td></tr><tr><td>orangemix</td><td>Abyss Orange Mix 3 (all variants)</td></tr><tr><td>randomix3</td><td>Randomix 3</td></tr><tr><td>refslave</td><td>RefSlave v2</td></tr><tr><td>vae-ft-mse-84000-ema-pruned</td><td>Alfamix, Experience 7.0, Realistic Experience, RPG v4, Simpmaker 3K, Stablydiffuseds Aesthetic Mix, Uber Realistic Porn Maker</td></tr></tbody></table><a href="https://i.ibb.co/61tqDn5/vae-grid-realistic.png">VAE comparison grid (realistic)</a> --- <a href="https://i.ibb.co/m6VqJH7/vae-grid-anime.png">VAE comparison grid (anime)</a></details>
Model = "526 Mix" #@param ["526 Mix", "7th Anime v3c", "8528 Diffusion", "Abyss 7th Layer", "A Certain Model", "A Certain Thing", "A Certainty", "Abyss Orange Mix 2 NSFW", "Abyss Orange Mix 2 HARD", "Abyss Orange Mix 3", "Abyss Orange Mix 3 A1", "Abyss Orange Mix 3 A1 B", "Abyss Orange Mix 3 A2", "Abyss Orange Mix 3 A3", "Alfamix", "All In One Pixel Model", "Analog Diffusion", "Anything V3", "Anything V4.5", "Anything Gape", "Arcane Diffusion 3", "Babes Blend", "Berry Mix 1.5", "Cafe Instagram v6", "Cetus Mix v3.1", "Chillout Mix", "Cinematic Diffusion", "Cocoa", "Color Bomb Mix", "Comic Diffusion v2", "Comics Blend", "Complex Lineart", "Cool Japan Diffusion 2.1", "Counterfeit v2.0", "Counterfeit v2.5 (pruned)", "Cyberpunk Anime Diffusion", "Defmix v2", "Deliberate", "Deliberate v2", "Dpepmkmp", "Dream Like Sam Kuvshino", "Dreamlike Diffusion 1", "Dreamlike Photoreal 2", "Dreamshaper 3.32 Baked vae (clip fix)", "DucHaitenAIArt v2.0", "Eimis Anime Diffusion v1", "Elldreth's Lucid Mix v10", "Elysium Anime v2", "Elysium Anime v3", "Experience 7.0", "F222", "Face Bomb Mix", "Faetastic v1", "Food Crit", "Grapefruit", "Graphic Art beta 11", "HasDX", "Hassan Blend 1.4", "Hyper Bomb Mix", "Illuminati v1.1", "Illustration Diffusion", "Inkpunk", "Izumi v1", "Knollingcase", "Liberty", "Magical Mix v1", "Magical Mix v2", "Mare Coloris", "Midjourney v4", "Mo Di Diffusion", "Monstermash 4 + Anything v3", "Monstermash 6 + Anything v3", "Mousey Mix", "Neverending Dream (baked vae)", "Open Anime Journey", "Open Journey v2", "Orange Cocoa Mix 2 HARD", "Padoru v1", "Papercut Diffusion", "Pastel Mix Stylized Anime (fp16)", "Pastel Mix Stylized Anime (fp32)", "Perfect World v2 (baked vae)", "Photorealostic Fuen v1", "Pony Diffusion 2", "Pony Diffusion 3", "Popup Book Diffusion", "Protogen v2.2", "Protogen X 3.4", "Protogen Nova", "PVC v3", "Rando mix 3", "Realbiter v1", "Realistic Experience", "Realistic Vision v1.3", "RefSlave v2", "Riffusion v1", "RPG v4", "SamDoesArt Diffusion", "SamDoesSexy Blend", "Seek Art Mega v1", "Simpmaker 3k", "Stable Diffusion 1.4", "Stable Diffusion 1.5", "Stable Diffusion 1.5 Inpainting", "Stable Diffusion 2", "Stable Diffusion 2 Base", "Stable Diffusion 2 Depth", "Stable Diffusion 2 Inpainting", "Stable Diffusion 2.1", "Stable Diffusion 2.1 Base", "Stablydiffuseds Aesthetic Mix", "Sukiyaki Mix", "Synthwave Punk v2", "Uber Realistic Porn Maker v1", "Van Gogh Diffusion", "Vinte Protogen Mix", "Vintedois Diffusion", "Voxo", "Waifu Diffusion 1.3", "Waifu Diffusion 1.4", "Waifu Diffusion 1.5 beta 1", "Waifu Diffusion 1.5 beta 2", "Waifu Diffusion 1.5 beta 2 Aesthetic", "Woolitize", "Yiffy E18", "Yiffy Mix", "Yohan Diffusion"]

#@markdown ##*Extensions*
#@markdown <details><summary>Click here to show/hide links to extensions' homepages</summary><a href="https://github.com/kohya-ss/sd-webui-additional-networks">Additional Networks</a> --- <a href="https://github.com/etherealxx/batchlinks-webui">Batchlinks Downloader</a> --- <a href="https://github.com/camenduru/sd-civitai-browser">CivitAI Browser</a> --- <a href="https://github.com/Mikubill/sd-webui-controlnet">ControlNet</a> --- <a href="https://github.com/deforum-art/deforum-for-automatic1111-webui">Deforum</a> --- <a href="https://github.com/jexom/sd-webui-depth-lib">Depth Map Library</a> --- <a href="https://github.com/camenduru/stable-diffusion-webui-huggingface">Huggingface</a> --- <a href="https://github.com/AlUlkesh/stable-diffusion-webui-images-browser">Images Browser</a> --- <a href="https://github.com/Klace/stable-diffusion-webui-instruct-pix2pix">Instruct Pix2Pix</a> --- <a href="https://github.com/camenduru/openpose-editor">OpenPose Editor</a> --- <a href="https://github.com/hnmr293/posex">PoseX</a> --- <a href="https://github.com/camenduru/stable-diffusion-webui-scripts">Run N Times</a></details>
additional_networks = True #@param {type:"boolean"}
batchlinks_downloader = False #@param {type:"boolean"}
civitai_browser = True #@param {type:"boolean"}
controlnet = False #@param {type:"boolean"}
deforum = False #@param {type:"boolean"}
depth_library = False #@param {type:"boolean"}
huggingface = False #@param {type:"boolean"}
images_browser = True #@param {type:"boolean"}
instruct_pix2pix_model = False #@param {type:"boolean"}
openpose_editor = False #@param {type:"boolean"}
posex = False #@param {type:"boolean"}
run_n_times_script = False #@param {type:"boolean"}

#@markdown ##*Negative embeddings*
bad_artist = True #@param {type:"boolean"} 
bad_artist_anime = True #@param {type:"boolean"}
bad_hands_5 = True #@param {type:"boolean"}
bad_prompt = True #@param {type:"boolean"}
bad_prompt_version2 = True #@param {type:"boolean"}
easynegative = True #@param {type:"boolean"}
unspeakable_horrors_64 = True #@param {type:"boolean"}
ng_deepnegative_v1_75t = True #@param {type:"boolean"}

#@markdown ##*VAEs*
anything_v3 = False #@param {type:"boolean"} 
anything_v4 = False #@param {type:"boolean"}
grapefruit = False #@param {type:"boolean"}
kl_f8_anime2 = False #@param {type:"boolean"}
orangemix = False #@param {type:"boolean"}
randomix_3 = False #@param {type:"boolean"}
refslave = False #@param {type:"boolean"}
vae_ft_mse_840000_ema_pruned = True #@param {type:"boolean"}

#@markdown ##*Choose a service to host the webui*
Tunnel = "Cloudflare" #@param ["Cloudflare", "Gradio", "Remote.moe"] 

#@markdown ##*Uncheck to launch webui without dark theme*
dark_theme = True #@param {type:'boolean'}

#####################
##### SHORTCUTS #####
#####################

root_folder = "/content/stable-diffusion-webui"
models_folder = "/content/stable-diffusion-webui/models/Stable-diffusion"
ctrlnet_url = "https://huggingface.co/ckpt/ControlNet/resolve/main"
ctrlnet_yaml_url = "https://huggingface.co/ckpt/ControlNet/raw/main"
ctrlnet_folder = "/content/stable-diffusion-webui/extensions/sd-webui-controlnet/models"
oppo_folder = "/content/stable-diffusion-webui/extensions/sd-webui-controlnet/annotator"
embeds_folder = "/content/stable-diffusion-webui/embeddings"
vaes_folder = "/content/stable-diffusion-webui/models/VAE"
aria_args = "-q -c -x 16 -s 16 -k 1M"

###################################
##### DICTIONNARIES AND LISTS #####
###################################

# Extensions repositories dictionnary

repos_dict = {
    "deforum-for-automatic1111-webui": {
        "repo": "https://github.com/deforum-art/deforum-for-automatic1111-webui",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": deforum
    },
    "stable-diffusion-webui-images-browser": {
        "repo": "https://github.com/AlUlkesh/stable-diffusion-webui-images-browser",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": images_browser
    },
    "stable-diffusion-webui-huggingface": {
        "repo": "https://github.com/camenduru/stable-diffusion-webui-huggingface",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": huggingface
    },
    "sd-civitai-browser": {
        "repo": "https://github.com/camenduru/sd-civitai-browser",
        "branch": "v2.0",
        "subfolder": "extensions",
        "command": "clone",
        "value": civitai_browser
    },
    "sd-webui-additional-networks": {
        "repo": "https://github.com/kohya-ss/sd-webui-additional-networks",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": additional_networks
    },
    "sd-webui-controlnet": {
        "repo": "https://github.com/Mikubill/sd-webui-controlnet",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": controlnet
    },
    "openpose-editor": {
        "repo": "https://github.com/camenduru/openpose-editor",
        "branch": "master",
        "subfolder": "extensions",
        "command": "clone",
        "value": openpose_editor
    },
    "batchlinks-webui": {
        "repo": "https://github.com/etherealxx/batchlinks-webui",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": batchlinks_downloader
    },
    "run_n_times.py": {
        "repo": "https://raw.githubusercontent.com/camenduru/stable-diffusion-webui-scripts/main/run_n_times.py",
        "branch": "",
        "subfolder": "scripts",
        "command": "aria2",
        "value": run_n_times_script
    },
    "instruct-pix2pix-00-22000.safetensors": {
        "repo": "https://huggingface.co/timbrooks/instruct-pix2pix/resolve/main/instruct-pix2pix-00-22000.safetensors",
        "branch": "",
        "subfolder": "models/Stable-diffusion",
        "command": "aria2",
        "value": instruct_pix2pix_model
    },
    "sd-webui-depth-lib": {
        "repo": "https://github.com/jexom/sd-webui-depth-lib",
        "branch": "main",
        "subfolder": "extensions",
        "command": "clone",
        "value": depth_library
    },
    "posex": {
        "repo": "https://github.com/hnmr293/posex",
        "branch": "master",
        "subfolder": "extensions",
        "command": "clone",
        "value": posex
    }
}

# ControlNet list

ctrlnet_list = [
    "control_canny-fp16.safetensors",
    "control_depth-fp16.safetensors",
    "control_hed-fp16.safetensors",
    "control_mlsd-fp16.safetensors",
    "control_normal-fp16.safetensors",
    "control_openpose-fp16.safetensors",
    "control_scribble-fp16.safetensors",
    "control_seg-fp16.safetensors",
    "t2iadapter_seg_sd14v1.pth",
    "t2iadapter_sketch_sd14v1.pth",
    "t2iadapter_seg_sd14v1.pth",
    "t2iadapter_openpose_sd14v1.pth",
    "t2iadapter_keypose_sd14v1.pth",
    "t2iadapter_depth_sd14v1.pth",
    "t2iadapter_color_sd14v1.pth",
    "t2iadapter_canny_sd14v1.pth",
    "canny-sd21-safe.safetensors",
    "depth-sd21-safe.safetensors",
    "hed-sd21-safe.safetensors",
    "openpose-sd21-safe.safetensors",
    "scribble-sd21-safe.safetensors"
]

ctrlnet_yaml_list = [
    "canny-sd21-safe.yaml",
    "depth-sd21-safe.yaml",
    "hed-sd21-safe.yaml",
    "openpose-sd21-safe.yaml",
    "scribble-sd21-safe.yaml"
]

# OpenPose dictionnary

oppo_dict = {
    "hand_pose_model.pth": "/openpose/",
    "body_pose_model.pth": "/openpose/",
    "dpt_hybrid-midas-501f0c75.pt": "/midas/",
    "mlsd_large_512_fp32.pth": "/mlsd/",
    "mlsd_tiny_512_fp32.pth": "/mlsd/",
    "network-bsds500.pth": "/hed/",
    "upernet_global_small.pth": "/uniformer/"
}

# Models dictionnary

models_dict = {
    "526 Mix": {
        "model_url": "https://civitai.com/api/download/models/19790",
        "model_file": "526mixExperimental_v12.safetensors"
    },
    "7th Anime v3c": {
        "model_url": "https://huggingface.co/syaimu/7th_Layer/resolve/main/7th_anime_v3/7th_anime_v3_C.safetensors",
        "model_file": "7th_anime_v3_C.safetensors"
    },
    "8528 Diffusion": {
        "model_url": "https://huggingface.co/ckpt/8528-diffusion/resolve/main/8528d-final.ckpt",
        "model_file": "8528d-final.ckpt"
    },
    "Abyss 7th Layer": {
        "model_url": "https://huggingface.co/syaimu/7th_Layer/resolve/main/7th_layer/Abyss_7th_layer.ckpt",
        "model_file": "Abyss_7th_layer.ckpt",
    },
    "A Certain Model": {
        "model_url": "https://huggingface.co/ckpt/ACertainModel/resolve/main/ACertainModel.ckpt",
        "model_file": "ACertainModel.ckpt"
    },
    "A Certain Thing": {
        "model_url": "https://huggingface.co/ckpt/ACertainThing/resolve/main/ACertainThing.ckpt",
        "model_file": "ACertainThing.ckpt"
    },
    "A Certainty": {
        "model_url": "https://huggingface.co/ckpt/ACertainty/resolve/main/ACertainty.ckpt",
        "model_file": "ACertainty.ckpt"
    },
    "Abyss Orange Mix 2 NSFW": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix2/AbyssOrangeMix2_nsfw.safetensors",
        "model_file": "AbyssOrangeMix2_nsfw.safetensors",
        "vae": "anything_v3"
    },
    "Abyss Orange Mix 2 HARD": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix2/AbyssOrangeMix2_hard.safetensors",
        "model_file": "AbyssOrangeMix2_hard.safetensors",
        "vae": "anything_v3"
    },
    "Abyss Orange Mix 3": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3_orangemixs.safetensors",
        "model_file": "AOM3_orangemixs.safetensors",
        "vae": "orangemix"
    },
    "Abyss Orange Mix 3 A1": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3A1_orangemixs.safetensors",
        "model_file": "AOM3A1_orangemixs.safetensors",
        "vae": "orangemix"
    },
    "Abyss Orange Mix 3 A1 B": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3A1B_orangemixs.safetensors",
        "model_file": "AOM3A1B_orangemixs.safetensors",
        "vae": "orangemix"
    },
    "Abyss Orange Mix 3 A2": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3A2_orangemixs.safetensors",
        "model_file": "AOM3A2_orangemixs.safetensors",
        "vae": "orangemix"
    },
    "Abyss Orange Mix 3 A3": {
        "model_url": "https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3A3_orangemixs.safetensors",
        "model_file": "AOM3A3_orangemixs.safetensors",
        "vae": "orangemix"
    },
    "Alfamix": {
        "model_url": "'https://civitai.com/api/download/models/17792?type=Model&format=SafeTensor'",
        "model_file": "alfamix_V1.safetensors",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "All In One Pixel Model": {
        "model_url": "https://huggingface.co/ckpt/All-In-dne-Pixel-Model/resolve/main/Public-Prompts-Pixel-Model.ckpt",
        "model_file": "Public-Prompts-Pixel-Model.ckpt"
    },
    "Analog Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Analog-Diffusion/resolve/main/analog-diffusion-1.0.ckpt",
        "model_file": "analog-diffusion-1.0.ckpt"
    },
    "Anything V3": {
        "model_url": "https://huggingface.co/ckpt/anything-v3.0/resolve/main/Anything-V3.0-pruned.ckpt",
        "model_file": "Anything-V3.0-pruned.ckpt",
        "vae": "anything_v3"
    },
    "Anything V4.5": {
        "model_url": "https://huggingface.co/ckpt/anything-v4.0/resolve/main/anything-v4.5-pruned.ckpt",
        "model_file": "anything-v4.5-pruned.ckpt",
        "vae": "anything_v4"
    },
    "Anything Gape": {
        "model_url": "https://huggingface.co/ckpt/AnyGape/resolve/main/AnythingGape-fp16.ckpt",
        "model_file": "AnythingGape-fp16.ckpt"
    },
    "Arcane Diffusion 3": {
        "model_url": "https://huggingface.co/ckpt/Arcane-Diffusion/resolve/main/arcane-diffusion-v3.ckpt",
        "model_file": "arcane-diffusion-v3.ckpt"
    },
    "Babes Blend": {
        "model_url": "https://huggingface.co/ckpt/babes/resolve/main/babes_11.safetensors",
        "model_file": "babes_11.safetensors"
    },
    "Berry Mix 1.5": {
        "model_url": "https://huggingface.co/ckpt/rbm/resolve/main/rbm.ckpt",
        "model_file": "rbm.ckpt"
    },
    "Cafe Instagram v6": {
        "model_url": "https://huggingface.co/ckpt/cafe-instagram-sd-1-5-v6/resolve/main/instagram-latest-plus-clip-v6e1_50000.safetensors",
        "model_file": "instagram-latest-plus-clip-v6e1_50000.safetensors"
    },
    "Cetus Mix v3.1": {
        "model_url": "https://civitai.com/api/download/models/20229",
        "model_file": "cetusMix_Codaedition.safetensors"
    },
    "Chillout Mix": {
        "model_url": "https://huggingface.co/ckpt/chilloutmix/resolve/main/chilloutmix_NiPrunedFp32Fix.safetensors",
        "model_file": "chilloutmix_NiPrunedFp32Fix.safetensors"
    },
    "Cinematic Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Cinematic-Diffusion/resolve/main/syberart.ckpt",
        "model_file": "syberart.ckpt"
    },
    "Cocoa": {
        "model_url": "https://huggingface.co/ckpt/yohan-diffusion/resolve/main/Cocoa.ckpt",
        "model_file": "Cocoa.ckpt",
        "vae": "anything_v3"
    },
    "Color Bomb Mix": {
        "model_url": "https://huggingface.co/mocker/KaBoom/resolve/main/ColorBombMix-fp16-no-ema.safetensors",
        "model_file": "ColorBombMix-fp16-no-ema.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Comic Diffusion v2": {
        "model_url": "https://huggingface.co/ckpt/Comic-Diffusion/resolve/main/comic-diffusion-V2.ckpt",
        "model_file": "comic-diffusion-V2.ckpt"
    },
    "Comics Blend": {
        "model_url": "https://huggingface.co/ckpt/comicsblend/resolve/main/comicsblend_V1.ckpt",
        "model_file": "comicsblend_V1.ckpt"
    },
    "Complex Lineart": {
        "model_url": "https://huggingface.co/ckpt/Complex-Lineart/resolve/main/ComplexLA%20Style.ckpt",
        "model_file": "ComplexLA_Style.ckpt"
    },
    "Cool Japan Diffusion 2.1": {
        "model_url": "https://huggingface.co/ckpt/cool-japan-diffusion-2-1-0/resolve/main/v2-1-0.ckpt",
        "model_file": "v2-1-0.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/cool-japan-diffusion-2-1-0/raw/main/v2-1-0.yaml",
        "yaml_file": "v2-1-0.yaml"
    },
    "Counterfeit v2.0": {
        "model_url": "https://huggingface.co/ckpt/Counterfeit-V2.0/resolve/main/Counterfeit-V2.0fp16.safetensors",
        "model_file": "Counterfeit-V2.0fp16.safetensors",
        "vae": "anything_v3"
    },
    "Counterfeit v2.5 (pruned)": {
        "model_url": "https://huggingface.co/gsdf/Counterfeit-V2.5/resolve/main/Counterfeit-V2.5_pruned.safetensors",
        "model_file": "Counterfeit-V2.5_pruned.safetensors",
        "vae": "anything_v3"
    },
    "Cyberpunk Anime Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Cyberpunk-Anime-Diffusion/resolve/main/Cyberpunk-Anime-Diffusion.ckpt",
        "model_file": "Cyberpunk-Anime-Diffusion.ckpt"
    },
    "Defmix v2": {
        "model_url": "https://huggingface.co/Defpoint/Defmix-v2.0/resolve/main/Defmix-v2.0.safetensors",
        "model_file": "Defmix-v2.0.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Deliberate": {
        "model_url": "https://huggingface.co/XpucT/Deliberate/resolve/main/Deliberate.safetensors",
        "model_file": "Deliberate.safetensors",
        "model_url": "https://huggingface.co/XpucT/Deliberate/resolve/main/Deliberate_v2.safetensors",
        "model_file": "Deliberate_v2.safetensors"
    },
    "Deliberate v2": {
        "model_url": "https://huggingface.co/XpucT/Deliberate/resolve/main/Deliberate_v2.safetensors",
        "model_file": "Deliberate_v2.safetensors"
    },
    "Dpepmkmp": {
        "model_url": "https://huggingface.co/closertodeath/dpepmkmp/resolve/main/dpepmkmp.safetensors",
        "model_file": "dpepmkmp.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Dream Like Sam Kuvshino": {
        "model_url": "https://huggingface.co/ckpt/dreamlikesamkuvshinov/resolve/main/dreamlikesamkuvshino.safetensors",
        "model_file": "dreamlikesamkuvshino.safetensors"
    },
    "Dreamlike Diffusion 1": {
        "model_url": "https://huggingface.co/ckpt/dreamlike-diffusion-1.0/resolve/main/dreamlike-diffusion-1.0.ckpt",
        "model_file": "dreamlike-diffusion-1.0.ckpt"
    },
    "Dreamlike Photoreal 2": {
        "model_url": "https://huggingface.co/ckpt/dreamlike-photoreal-2.0/resolve/main/dreamlike-photoreal-2.0.safetensors",
        "model_file": "dreamlike-photoreal-2.0.safetensors"
    },
    "Dreamshaper 3.32 Baked vae (clip fix)": {
        "model_url": "https://huggingface.co/Lykon/DreamShaper/resolve/main/Dreamshaper_3.32_baked_vae_clip_fix.safetensors",
        "model_file": "Dreamshaper_3.32_baked_vae_clip_fix.safetensors"
    },
    "DucHaitenAIArt v2.0": {
        "model_url": "https://huggingface.co/DucHaiten/DucHaitenAIart/resolve/main/DucHaitenAIart_v2.0.safetensors",
        "model_file": "DucHaitenAIart_v2.0.safetensors"
    },
    "Eimis Anime Diffusion v1": {
        "model_url": "https://huggingface.co/ckpt/EimisAnimeDiffusion_1.0v/resolve/main/EimisAnimeDiffusion_1-0v.ckpt",
        "model_file": "EimisAnimeDiffusion_1-0v.ckpt"
    },
    "Elldreth's Lucid Mix v10": {
        "model_url": "https://huggingface.co/ckpt/elldrethSLucidMix_v10/resolve/main/elldrethSLucidMix_v10.ckpt",
        "model_file": "elldrethSLucidMix_v10.ckpt"
    },
    "Elysium Anime v2": {
        "model_url": "https://huggingface.co/ckpt/SD-Elysium-Model/resolve/main/Elysium_Anime_V2.ckpt",
        "model_file": "Elysium_Anime_V2.ckpt",
        "vae": "kl_f8_anime2"
    },
    "Elysium Anime v3": {
        "model_url": "https://huggingface.co/ckpt/SD-Elysium-Model/resolve/main/Elysium_Anime_V3.safetensors",
        "model_file": "Elysium_Anime_V3.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Experience 7.0": {
        "model_url": "https://huggingface.co/Schisim/Experience/resolve/main/eXp_7.0.safetensors",
        "model_file": "eXp_7.0.safetensors",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "F222": {
        "model_url": "https://huggingface.co/ckpt/f2/resolve/main/f2.ckpt",
        "model_file": "f2.ckpt"
    },
    "Face Bomb Mix": {
        "model_url": "https://huggingface.co/mocker/KaBoom/resolve/main/FaceBombMix-fp16-no-ema.safetensors",
        "model_file": "FaceBombMix-fp16-no-ema.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Faetastic v1": {
        "model_url": "'https://civitai.com/api/download/models/16553?type=Model&format=SafeTensor'",
        "model_file": "faetastic_.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Food Crit": {
        "model_url": "https://huggingface.co/ckpt/food-crit/resolve/main/food_crit.ckpt",
        "model_file": "food_crit.ckpt"
    },
    "Grapefruit": {
        "model_url": "https://huggingface.co/ckpt/grapefruit/resolve/main/grapefruit.safetensors",
        "model_file": "grapefruit.safetensors",
        "vae": "grapefruit"
    },
    "Graphic Art beta 11": {
        "model_url": "'https://civitai.com/api/download/models/19546?type=Model&format=SafeTensor'",
        "model_file": "graphicArt_graphicArtBeta11.safetensors",
        "yaml_url": "'https://civitai.com/api/download/models/19546?type=Config&format=Other'",
        "yaml_file": "graphicArt_graphicArtBeta11.yaml"
    },
    "HasDX": {
        "model_url": "https://huggingface.co/ckpt/hasdx/resolve/main/hasdx_emaonly.ckpt",
        "model_file": "hasdx_emaonly.ckpt"
    },
    "Hassan Blend 1.4": {
        "model_url": "https://huggingface.co/ckpt/hassanblend1.4/resolve/main/HassanBlend1.4-Pruned.ckpt",
        "model_file": "HassanBlend1.4-Pruned.ckpt"
    },
    "Hyper Bomb Mix": {
        "model_url": "https://huggingface.co/mocker/KaBoom/resolve/main/HyperBombMix-fp16-no-ema.safetensors",
        "model_file": "HyperBombMix-fp16-no-ema.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Illuminati v1.1": {
        "model_url": "https://civitai.com/api/download/models/13259",
        "model_file": "illuminatiDiffusionV1_v11.safetensors",
        "yaml_url": "https://huggingface.co/IlluminatiAI/Illuminati_Diffusion_v1.0/raw/main/illuminati_diffusion_v1.0.yaml",
        "yaml_file": "illuminatiDiffusionV1_v11.yaml",
        "embeds_lists": {
            "embeds_urls": [
                "https://civitai.com/api/download/models/15927",
                "https://civitai.com/api/download/models/15926",
                "https://civitai.com/api/download/models/15921"
            ],
            "embeds_files": [
                "nrealfixer.pt",
                "nartfixer.pt",
                "nfixer.pt"
            ]
        }        
    },
    "Illustration Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Illustration-Diffusion/resolve/main/hollie-mengert.ckpt",
        "model_file": "hollie-mengert.ckpt"
    },
    "Inkpunk": {
        "model_url": "https://huggingface.co/ckpt/Inkpunk-Diffusion/resolve/main/Inkpunk-Diffusion-v2.ckpt",
        "model_file": "Inkpunk-Diffusion-v2.ckpt"
    },
    "Izumi v1": {
        "model_url": "https://huggingface.co/ckpt/iz/resolve/main/izumi_01.ckpt",
        "model_file": "izumi_01.ckpt"
    },
    "Knollingcase": {
        "model_url": "https://huggingface.co/ckpt/knollingcase/resolve/main/knollingcase.ckpt",
        "model_file": "knollingcase.ckpt"
    },
    "Liberty": {
        "model_url": "'https://civitai.com/api/download/models/6908?type=Model&format=SafeTensor'",
        "model_file": "liberty_main.safetensors"
    },
    "Magical Mix v1": {
        "model_url": "https://huggingface.co/ckpt/MagicalMix_v1/resolve/main/MagicalMix_v1.safetensors",
        "model_file": "MagicalMix_v1.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Magical Mix v2": {
        "model_url": "https://huggingface.co/mekabu/MagicalMix_v2/resolve/main/MagicalMix_v2.safetensors",
        "model_file": "MagicalMix_v2.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Mare Coloris": {
        "model_url": "https://huggingface.co/ckpt/MareColoris/resolve/main/MareColoris.safetensors",
        "model_file": "MareColoris.safetensors"
    },
    "Midjourney v4": {
        "model_url": "https://huggingface.co/ckpt/midjourney-v4-diffusion/resolve/main/mdjrny-v4.ckpt",
        "model_file": "mdjrny-v4.ckpt"
    },
    "Mo Di Diffusion": {
        "model_url": "https://huggingface.co/ckpt/mo-di-diffusion/resolve/main/moDi-v1-pruned.ckpt",
        "model_file": "moDi-v1-pruned.ckpt"
    },
    "Monstermash 4 + Anything v3": {
        "model_url": "https://huggingface.co/ckpt/monstermash_anyv3/resolve/main/monstermash4%2Banyv3.safetensors",
        "model_file": "monstermash4%2Banyv3.safetensors",
        "vae": "anything_v3"
    },
    "Monstermash 6 + Anything v3": {
        "model_url": "https://huggingface.co/ckpt/monstermash_anyv3/resolve/main/monstermash6%2Banyv3.safetensors",
        "model_file": "monstermash6%2Banyv3.safetensors",
        "vae": "anything_v3"
    },
    "Mousey Mix": {
        "model_url": "https://huggingface.co/closertodeath/mouseymix/resolve/main/mouseymix.safetensors",
        "model_file": "mouseymix.safetensors"
    },
    "Neverending Dream (baked vae)": {
        "model_url": "https://huggingface.co/Lykon/NeverEnding-Dream/resolve/main/NeverEndingDream_ft_mse.safetensors",
        "model_file": "neverending_dream_bakedvae.safetensors"
    },
    "Open Anime Journey": {
        "model_url": "https://huggingface.co/ckpt/OpenAnimeJourney/resolve/main/OpenAnimeJourney.ckpt",
        "model_file": "OpenAnimeJourney.ckpt",
        "vae": "anything_v3"
    },
    "Open Journey v2": {
        "model_url": "https://huggingface.co/ckpt/openjourney-v2/resolve/main/openjourney-v2.ckpt",
        "model_file": "openjourney-v2.ckpt"
    },
    "Orange Cocoa Mix 2 HARD": {
        "model_url": "https://huggingface.co/ckpt/OrangeCocoaMix/resolve/main/OrangeCocoaMix2_hard.safetensors",
        "model_file": "OrangeCocoaMix2_hard.safetensors",
        "vae": "anything_v3"
    },
    "Padoru v1": {
        "model_url": "https://civitai.com/api/download/models/1522",
        "model_file": "padoruDiffusion_v1.ckpt",
    },
    "Papercut Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Stable_Diffusion_PaperCut_Model/resolve/main/PaperCut_v1.ckpt",
        "model_file": "PaperCut_v1.ckpt"
    },
    "Pastel Mix Stylized Anime (fp16)": {
        "model_url": "https://huggingface.co/andite/pastel-mix/resolve/main/pastelmix-fp16.ckpt",
        "model_file": "pastelmix-fp16.ckpt",
        "vae": "kl_f8_anime2",
    },
    "Pastel Mix Stylized Anime (fp32)": {
        "model_url": "https://huggingface.co/andite/pastel-mix/resolve/main/pastelmix-fp32.ckpt",
        "model_file": "pastelmix-fp32.ckpt",
        "vae": "kl_f8_anime2"
    },
    "Perfect World v2 (baked vae)": {
        "model_url": "https://huggingface.co/ckpt/perfect_world/resolve/main/perfectWorld_v2Baked.safetensors",
        "model_file": "perfectWorld_v2Baked.safetensors"
    },
    "Photorealostic Fuen v1": {
        "model_url": "https://huggingface.co/ckpt/photorealistic-fuen-v1/resolve/main/photorealistic-fuen-v1.ckpt",
        "model_file": "photorealistic-fuen-v1.ckpt"
    },
    "Pony Diffusion 2": {
        "model_url": "https://huggingface.co/ckpt/pony-diffusion-v2/resolve/main/pony-diffusion-v2.ckpt",
        "model_file": "pony-diffusion-v2.ckpt"
    },
    "Pony Diffusion 3": {
        "model_url": "https://huggingface.co/ckpt/pony-diffusion-v3/resolve/main/pony-diffusion-v3.ckpt",
        "model_file": "pony-diffusion-v3.ckpt"
    },
    "Popup Book Diffusion": {
        "model_url": "https://huggingface.co/ckpt/popupBook-diffusion/resolve/main/popupBook-diffusion-v1.ckpt",
        "model_file": "popupBook-diffusion-v1.ckpt"
    },
    "Protogen v2.2": {
        "model_url": "https://huggingface.co/ckpt/Protogen_V2.2/resolve/main/Protogen_V2.2.ckpt",
        "model_file": "Protogen_V2.2.ckpt"
    },
    "Protogen X 3.4": {
        "model_url": "https://huggingface.co/darkstorm2150/Protogen_x3.4_Official_Release/resolve/main/ProtoGen_X3.4.safetensors",
        "model_file": "Protogen_X3.4.safetensors"
    },
    "Protogen Nova": {
        "model_url": "https://huggingface.co/ckpt/Protogen_Nova_Official_Release/resolve/main/Protogen_Nova_Official_Release.ckpt",
        "model_file": "Protogen_Nova_Official_Release.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/Protogen_Nova_Official_Release/raw/main/Protogen_Nova_Official_Release.yaml",
        "yaml_file": "Protogen_Nova_Official_Release.yaml"
    },
    "PVC v3": {
        "model_url": "https://huggingface.co/p1atdev/pvc-v3/resolve/checkpoints/pvc-v3-fp32.safetensors",
        "model_file": "pvc-v3-fp32.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Rando mix 3": {
        "model_url": "https://huggingface.co/ckpt/RandoMix3/resolve/main/RandoMix3.ckpt",
        "model_file": "RandoMix3.ckpt",
        "vae": "randomix_3"
    },
    "Realbiter v1": {
        "model_url": "'https://civitai.com/api/download/models/19590?type=Model&format=SafeTensor'",
        "model_file": "realbiter_v10.safetensors"
    },
    "Realistic Experience": {
        "model_url": "https://huggingface.co/Schisim/Experience/resolve/main/Realistic_EXP.safetensors",
        "model_file": "Realistic_EXP.safetensors",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "Realistic Vision v1.3": {
        "model_url": "'https://civitai.com/api/download/models/6987?type=Model&format=SafeTensor'",
        "model_file": "realisticVisionV13_v13.safetensors"
    },
    "RefSlave v2": {
        "model_url": "https://huggingface.co/ckpt/bucket/resolve/main/refslaveV2_v2-emaonly.safetensors",
        "model_file": "RefSlave-V2.safetensors",
        "vae": "refslave"
    },
    "Riffusion v1": {
        "model_url": "https://huggingface.co/ckpt/riffusion-model-v1/resolve/main/riffusion-model-v1.ckpt",
        "model_file": "riffusion-model-v1.ckpt"
    },
    "RPG v4": {
        "model_url": "https://huggingface.co/Anashel/rpg/resolve/main/RPG-V4-Model-Download/RPG-v4.safetensors",
        "model_file": "RPG-v4.safetensors",
        "yaml_url": "https://huggingface.co/Anashel/rpg/raw/main/RPG-V4-Model-Download/RPG-v4.yaml",
        "yaml_file": "RPG-v4.yaml",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "SamDoesArt Diffusion": {
        "model_url": "https://huggingface.co/ckpt/SamDoesArt-V3/resolve/main/SamDoesArt-V3.ckpt",
        "model_file": "SamDoesArt-V3.ckpt"
    },
    "SamDoesSexy Blend": {
        "model_url": "https://huggingface.co/ckpt/samdoessexy-blend/resolve/main/samdoessexyBlend.safetensors",
        "model_file": "samdoessexyBlend.safetensors"
    },
    "Seek Art Mega v1": {
        "model_url": "https://huggingface.co/ckpt/seek.art_MEGA/resolve/main/seek_art_mega_v1.ckpt",
        "model_file": "seek_art_mega_v1.ckpt"
    },
    "Simpmaker 3k": {
        "model_url": "https://huggingface.co/ckpt/SimpMaker-3K1/resolve/main/SimpMaker%203k.ckpt",
        "model_file": "SimpMaker%203k.ckpt",
        "vae": "vae_ft_mse_840000_ema_pruned",
    },
    "Stable Diffusion 1.4": {
        "model_url": "https://huggingface.co/ckpt/sd14/resolve/main/sd-v1-4.ckpt",
        "model_file": "sd-v1-4.ckpt"
    },
    "Stable Diffusion 1.5": {
        "model_url": "https://huggingface.co/ckpt/sd15/resolve/main/v1-5-pruned-emaonly.ckpt",
        "model_file": "v1-5-pruned-emaonly.ckpt"
    },
    "Stable Diffusion 1.5 Inpainting": {
        "model_url": "https://huggingface.co/ckpt/i15/resolve/main/sd-v1-5-inpainting.ckpt",
        "model_file": "sd-v1-5-inpainting.ckpt"
    },
    "Stable Diffusion 2": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2/resolve/main/768-v-ema.ckpt",
        "model_file": "768-v-ema.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2/raw/main/v2-inference-v.yaml",
        "yaml_file": "768-v-ema.yaml"
    },
    "Stable Diffusion 2 Base": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2-base/resolve/main/512-base-ema.ckpt",
        "model_file": "512-base-ema.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2-base/raw/main/v2-inference.yaml",
        "yaml_file": "512-base-ema.yaml"
    },
    "Stable Diffusion 2 Depth": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2-depth/resolve/main/512-depth-ema.ckpt",
        "model_file": "512-depth-ema.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2-depth/raw/main/v2-midas-inference.yaml",
        "yaml_file": "512-depth-ema.yaml",
        "embeds_lists": {
            "embeds_urls": [
                "https://huggingface.co/spaablauw/FloralMarble/resolve/main/FloralMarble-400.pt",
            ],
            "embeds_files": [
                "FloralMarble-400.pt"
            ]
        }
    },
    "Stable Diffusion 2 Inpainting": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2-inpainting/resolve/main/512-inpainting-ema.ckpt",
        "model_file": "512-inpainting-ema.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2-inpainting/raw/main/v2-inpainting-inference.yaml",
        "yaml_file": "512-inpainting-ema.yaml"
    },
    "Stable Diffusion 2.1": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2-1/resolve/main/v2-1_768-ema-pruned.ckpt",
        "model_file": "v2-1_768-ema-pruned.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2-1/raw/main/v2-inference-v.yaml",
        "yaml_file": "v2-1_768-ema-pruned.yaml",
        "embeds_lists": {
            "embeds_urls": [
                "https://huggingface.co/joachimsallstrom/Double-Exposure-Embedding/resolve/main/dblx.pt",
                "https://huggingface.co/Conflictx/AnimeScreencap/resolve/main/AnimeScreenCap.pt"
            ],
            "embeds_files": [
                "dblx.pt",
                "AnimeScreenCap.pt"
            ]
        }
    },
    "Stable Diffusion 2.1 Base": {
        "model_url": "https://huggingface.co/ckpt/stable-diffusion-2-1-base/resolve/main/v2-1_512-ema-pruned.ckpt",
        "model_file": "v2-1_512-ema-pruned.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/stable-diffusion-2-1-base/raw/main/v2-inference.yaml",
        "yaml_file": "v2-1_512-ema-pruned.yaml"
    },
    "Stablydiffuseds Aesthetic Mix": {
        "model_url": "'https://civitai.com/api/download/models/6266?type=Pruned%20Model&format=SafeTensor'",
        "model_file": "stablydiffuseds_26.safetensors",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "Sukiyaki Mix": {
        "model_url": "https://huggingface.co/Vsukiyaki/SukiyakiMix-v1.0/resolve/main/SukiyakiMix-v1.0-fp32.safetensors",
        "model_file": "SukiyakiMix-v1.0"
    },
    "Synthwave Punk v2": {
        "model_url": "https://huggingface.co/ckpt/SynthwavePunk-v2/resolve/main/SynthwavePunk-v2.ckpt",
        "model_file": "SynthwavePunk-v2.ckpt"
    },
    "Uber Realistic Porn Maker v1": {
        "model_url": "https://huggingface.co/ckpt/urp/resolve/main/uberRealisticPornMer_urpMv1_ema.ckpt",
        "model_file": "uberRealisticPornMer_urpMv1_ema.ckpt",
        "vae": "vae_ft_mse_840000_ema_pruned"
    },
    "Van Gogh Diffusion": {
        "model_url": "https://huggingface.co/ckpt/Van-Gogh-diffusion/resolve/main/Van-Gogh-Style-lvngvncnt-v2.ckpt",
        "model_file": "Van-Gogh-Style-lvngvncnt-v2.ckpt"
    },
    "Vinte Protogen Mix": {
        "model_url": "https://civitai.com/api/download/models/6582",
        "model_file": "vinteprotogenmix_v10.safetensors"
    },
    "Vintedois Diffusion": {
        "model_url": "https://huggingface.co/ckpt/vintedois-diffusion-v0-1/resolve/main/vintedois_0_1.ckpt",
        "model_file": "vintedois_0_1.ckpt"
    },
    "Voxo": {
        "model_url": "https://huggingface.co/junjuice0/VOXO/resolve/main/VOXO-v0.ckpt",
        "model_file": "VOXO-v0.ckpt"
    },
    "Waifu Diffusion 1.3": {
        "model_url": "https://huggingface.co/ckpt/waifu-diffusion-v1-3/resolve/main/wd-v1-3-float32.ckpt",
        "model_file": "wd-v1-3-float32.ckpt",
        "vae": "kl_f8_anime2"
    },
    "Waifu Diffusion 1.4": {
        "model_url": "https://huggingface.co/ckpt/waifu-diffusion-v1-4/resolve/main/wd-1-4-anime_e1.ckpt",
        "model_file": "wd-1-4-anime_e1.ckpt",
        "yaml_url": "https://huggingface.co/ckpt/waifu-diffusion-v1-4/raw/main/wd-1-4-anime_e1.yaml",
        "yaml_file": "wd-1-4-anime_e1.yaml",
        "vae": "kl_f8_anime2"
    },
    "Waifu Diffusion 1.5 beta 1": {
        "model_url": "https://huggingface.co/ckpt/wd-1-5-beta/resolve/main/wd15-beta1-fp32.safetensors",
        "model_file": "wd15-beta1-fp32.safetensors",
        "yaml_url": "https://huggingface.co/ckpt/wd-1-5-beta/raw/main/wd15-beta1-fp32.yaml",
        "yaml_file": "wd15-beta1-fp32.yaml",
        "vae": "kl_f8_anime2"
    },
    "Waifu Diffusion 1.5 beta 2": {
        "model_url": "https://huggingface.co/waifu-diffusion/wd-1-5-beta2/resolve/main/checkpoints/wd-1-5-beta2-fp32.safetensors",
        "model_file": "wd-1-5-beta2-fp32.safetensors",
        "yaml_url": "https://huggingface.co/waifu-diffusion/wd-1-5-beta2/raw/main/checkpoints/wd-1-5-beta2-fp32.yaml",
        "yaml_file": "wd-1-5-beta2-fp32.yaml",
        "vae": "kl_f8_anime2"
    },
    "Waifu Diffusion 1.5 beta 2 Aesthetic": {
        "model_url": "https://huggingface.co/waifu-diffusion/wd-1-5-beta2/resolve/main/checkpoints/wd-1-5-beta2-aesthetic-fp32.safetensors",
        "model_file": "wd-1-5-beta2-aesthetic-fp32.safetensors",
        "yaml_url": "https://huggingface.co/waifu-diffusion/wd-1-5-beta2/raw/main/checkpoints/wd-1-5-beta2-aesthetic-fp32.yaml",
        "yaml_file": "wd-1-5-beta2-aesthetic-fp32.yaml",
        "vae": "kl_f8_anime2"
    },
    "Woolitize": {
        "model_url": "https://huggingface.co/ckpt/woolitize/resolve/main/woolitize.ckpt",
        "model_file": "woolitize.ckpt"
    },
    "Yiffy E18": {
        "model_url": "https://huggingface.co/ckpt/yiffy-e18/resolve/main/yiffy-e18.ckpt",
        "model_file": "yiffy-e18.ckpt"
    },
    "Yiffy Mix": {
        "url1": "https://huggingface.co/ckpt/yiffymix/resolve/main/yiffymix.safetensors",
        "file1": "yiffymix.safetensors",
        "vae": "kl_f8_anime2"
    },
    "Yohan Diffusion": {
        "model_url": "https://huggingface.co/ckpt/yohan-diffusion/resolve/main/yohan-diffusion.ckpt",
        "model_file": "yohan-diffusion.ckpt",
        "vae": "anything_v3"
    }
}

# Embeddings dictionnary

embeds_dict = {
    "bad_artist": {
        "embed_url": "https://huggingface.co/nick-x-hacker/bad-artist/resolve/main/bad-artist.pt",
        "embed_file": "bad-artist.pt",
        "value": bad_artist
    },
    "bad_artist_anime": {
        "embed_url": "https://huggingface.co/nick-x-hacker/bad-artist/resolve/main/bad-artist-anime.pt",
        "embed_file": "bad-artist-anime.pt",
        "value": bad_artist_anime
    },
    "bad_hands_5": {
        "embed_url": "https://huggingface.co/yesyeahvh/bad-hands-5/resolve/main/bad-hands-5.pt",
        "embed_file": "bad-hands-5.pt",
        "value": bad_hands_5
    },
    "bad_prompt": {
        "embed_url": "https://huggingface.co/datasets/Nerfgun3/bad_prompt/resolve/main/bad_prompt.pt",
        "embed_file": "bad_prompt.pt",
        "value": bad_prompt
    },
    "bad_prompt_version2": {
        "embed_url": "https://huggingface.co/datasets/Nerfgun3/bad_prompt/resolve/main/bad_prompt_version2.pt",
        "embed_file": "bad_prompt_version2.pt",
        "value": bad_prompt_version2
    },
    "easynegative": {
        "embed_url": "https://civitai.com/api/download/models/9536",
        "embed_file": "easynegative.pt",
        "value": easynegative
    },
    "ng_deepnegative_v1_75t": {
        "embed_url": "https://civitai.com/api/download/models/5637",
        "embed_file": "ng_deepnegative_v1_75t.pt",
        "value": ng_deepnegative_v1_75t
    },
    "unspeakable_horrors_64": {
        "embed_url": "https://civitai.com/api/download/models/5103",
        "embed_file": "Unspeakable-Horrors-64v.pt",
        "value": unspeakable_horrors_64
    },
}

# VAEs dictionnary

vaes_dict = {
    "anything_v3": {
        "vae_url": "https://huggingface.co/ckpt/anything-v3.0/resolve/main/Anything-V3.0.vae.pt",
        "vae_file": "Anything-V3.0.vae.pt",
        "value": anything_v3
    },
    "anything_v4": {
        "vae_url": "https://huggingface.co/ckpt/anything-v4.0/resolve/main/anything-v4.0.vae.pt",
        "vae_file": "anything-v4.0.vae.pt",
        "value": anything_v4
    },
    "grapefruit": {
        "vae_url": "https://huggingface.co/ckpt/grapefruit/resolve/main/grapefruit.vae.pt",
        "vae_file": "grapefruit.vae.pt",
        "value": grapefruit
    },
    "kl_f8_anime2": {
        "vae_url": "https://huggingface.co/hakurei/waifu-diffusion-v1-4/resolve/main/vae/kl-f8-anime2.ckpt",
        "vae_file": "kl-f8-anime2.ckpt",
        "value": kl_f8_anime2
    },
    "orangemix": {
        "vae_url": "https://huggingface.co/ckpt/OrangeMixs/resolve/main/orangemix.vae.pt",
        "vae_file": "orangemix.vae.pt",
        "value": orangemix
    },
    "randomix_3": {
        "vae_url": "https://huggingface.co/ckpt/RandoMix3/resolve/main/RandoMix3.vae.pt",
        "vae_file": "RandoMix3.vae.pt",
        "value": randomix_3
    },
    "refslave": {
        "vae_url": "https://huggingface.co/ckpt/bucket/resolve/main/RefSlave-vae.pt",
        "vae_file": "RefSlave-vae.pt",
        "value": refslave
    },
    "vae_ft_mse_840000_ema_pruned": {
        "vae_url": "https://huggingface.co/stabilityai/sd-vae-ft-mse-original/resolve/main/vae-ft-mse-840000-ema-pruned.safetensors",
        "vae_file": "vae-ft-mse-840000-ema-pruned.safetensors",
        "value": vae_ft_mse_840000_ema_pruned
    }
}

# Tunnels dictionnary

tunnels = {
    "Cloudflare": "--cloudflared",
    "Gradio": "",
    "Remote.moe": "--remotemoe"
}

################
##### CODE #####
################

# Installing requirements
print("Moving to:")
%cd /content
print("Installing aria2...")
!apt-get -y install aria2 > /dev/null
print("Installing xformers (pre-release)...")
!pip install -q --pre xformers==0.0.17.dev476 -U
print("Installing triton (pre-release)...")
!pip install -q --pre triton

# Daviljoe193 ram management fix

print("Setting up Daviljoe193's fix for colab's ram management...")
!aria2c $aria_args http://launchpadlibrarian.net/367274644/libgoogle-perftools-dev_2.5-2.2ubuntu3_amd64.deb
!aria2c $aria_args https://launchpad.net/ubuntu/+source/google-perftools/2.5-2.2ubuntu3/+build/14795286/+files/google-perftools_2.5-2.2ubuntu3_all.deb
!aria2c $aria_args https://launchpad.net/ubuntu/+source/google-perftools/2.5-2.2ubuntu3/+build/14795286/+files/libtcmalloc-minimal4_2.5-2.2ubuntu3_amd64.deb
!aria2c $aria_args https://launchpad.net/ubuntu/+source/google-perftools/2.5-2.2ubuntu3/+build/14795286/+files/libgoogle-perftools4_2.5-2.2ubuntu3_amd64.deb
!apt -y install libunwind8-dev > /dev/null 2>&1
!dpkg -i *.deb > /dev/null 2>&1
%env LD_PRELOAD=libtcmalloc.so
!rm *.deb

# Downloading main Stable Diffusion repository

print("Downloading Camenduru's stable-diffusion-webui repository from v2.1 branch")
!git clone -q -b v2.1 https://github.com/camenduru/stable-diffusion-webui
print("Downloading Tunnels extension")
!git clone -q -b main https://github.com/camenduru/sd-webui-tunnels /content/stable-diffusion-webui/extensions/sd-webui-tunnels

# Downloading requested extensions' repositories and models

for name, data in repos_dict.items():
    if data["value"]:
        clone_path = f"{root_folder}/{data['subfolder']}/{name}"
        aria2_path = f"{root_folder}/{data['subfolder']}"
        if data["command"] == "clone":
            print(f"Downloading extension : {name}")
            !git clone -q -b {data["branch"]} {data["repo"]} {clone_path}
        elif data["command"] == "aria2":
            print(f"Downloading extension model or file : {name}")
            !aria2c $aria_args {data["repo"]} -d {aria2_path} -o {name}

# Downloading models for ControlNet

if controlnet:
    for ctrlnet_model in ctrlnet_list:
        print (f"Downloading ControlNet model : {ctrlnet_model}")
        !aria2c $aria_args {ctrlnet_url}/{ctrlnet_model} -d {ctrlnet_folder} -o {ctrlnet_model}
    for ctrlnet_yaml in ctrlnet_yaml_list:
        print (f"Downloading ControlNet config : {ctrlnet_yaml}")
        !aria2c $aria_args {ctrlnet_yaml_url}/{ctrlnet_yaml} -d {ctrlnet_folder} -o {ctrlnet_yaml}
    for filename, path in oppo_dict.items():
        print(f"Downloading OpenPose model : {filename}")
        !aria2c $aria_args {ctrlnet_url}/{filename} -d {oppo_folder}{path} -o {filename}

# Downloading selected model

if Model in models_dict:
    model_url = models_dict[Model]["model_url"]
    model_file = models_dict[Model]["model_file"]
    print(f"Downloading Stable Diffusion model file : {model_file}")
    !aria2c $aria_args {model_url} -d {models_folder} -o {model_file}
    ckpt_arg = f"{models_folder}/{model_file}"
    if "yaml_url" in models_dict[Model]:
        yaml_url = models_dict[Model]["yaml_url"]
        yaml_file = models_dict[Model]["yaml_file"]
        print(f"Downloading {Model} yaml file : {yaml_file}")
        !aria2c $aria_args {yaml_url} -d {models_folder} -o {yaml_file}
    if "embeds_lists" in models_dict[Model]:
        embeds_urls = models_dict[Model]["embeds_lists"]["embeds_urls"]
        embeds_files = models_dict[Model]["embeds_lists"]["embeds_files"]
        for i, embed_url in enumerate(embeds_urls):
            embed_file = embeds_files[i]
            print(f"Downloading {Model} embedding (textual inversion) {i+1}: {embed_file}")
            !aria2c $aria_args {embed_url} -d {embeds_folder} -o {embed_file}
    if "vae" in models_dict[Model]:
        vae_name = models_dict[Model]["vae"]
        if vaes_dict[vae_name]["value"] == False:
            vaes_dict[vae_name]["value"] = True
        print(f"VAE {vae_name} will be downloaded as recommended by this model.")
    if Model == "Stable diffusion 2 Depth":
        print("Adding 'name = list(data.keys())[0]' to 'textual_inversion.py'")
        !sed -i -e '/emb = emb.unsqueeze(0)/a\                name = list(data.keys())[0]' /content/stable-diffusion-webui/modules/textual_inversion/textual_inversion.py
    if Model == "Riffusion v1":
        print("Downloading extension : sd-webui-riffusion")
        !git clone -q https://github.com/enlyth/sd-webui-riffusion /content/stable-diffusion-webui/extensions/sd-webui-riffusion
        print("Deleting install.py and requirements.txt from the extension folder")
        !rm /content/stable-diffusion-webui/extensions/sd-webui-riffusion/install.py /content/stable-diffusion-webui/extensions/sd-webui-riffusion/requirements.txt
        print("Installing package pedalboard")
        !pip -q install pedalboard
        
# Downloading selected embeddings

for embed_name, embed_info in embeds_dict.items():
    if embed_info["value"]:
        embed_url = embed_info["embed_url"]
        embed_file = embed_info["embed_file"]
        print(f"Downloading embedding (textual inversion): {embed_file}")
        !aria2c $aria_args {embed_url} -d {embeds_folder} -o {embed_file}

# Downloading selected VAEs

for vae_name, vae_info in vaes_dict.items():
    if vae_info["value"]:
        vae_url = vae_info["vae_url"]
        vae_file = vae_info["vae_file"]
        print(f"Downloading VAE: {vae_file}")
        !aria2c $aria_args {vae_url} -d {vaes_folder} -o {vae_file}

# Adding tunnel argument

for tunnel_key, tunnel_value in tunnels.items():
    if Tunnel == tunnel_key:
        host = tunnel_value

# Setting dark theme on/off

if dark_theme:
    dark = "--theme dark"

####################
##### HOTFIXES #####
####################

print("Hotfixes :")

## ControlNet/Deforum "folder not found"

if deforum and controlnet:
    print("Fixing Deforum x ControlNet 'folder not found' error")
    !mkdir /content/stable-diffusion-webui/extensions/deforum-for-automatic1111-webui/models

## Fixing some launcher stuff

print("Changing 'dict()))' to 'dict())).cuda()' in util.py")
!sed -i -e '''/    prepare_environment()/a\    os.system\(f\"""sed -i -e ''\"s/dict()))/dict())).cuda()/g\"'' /content/stable-diffusion-webui/repositories/stable-diffusion-stability-ai/ldm/util.py""")''' /content/stable-diffusion-webui/launch.py
print("Changing fastapi version requirement from 0.90.1 to 0.89.1")
!sed -i -e 's/fastapi==0.90.1/fastapi==0.89.1/g' /content/stable-diffusion-webui/requirements_versions.txt

###########################
##### LAUNCHING WEBUI #####
###########################

print("Moving to :")
%cd /content/stable-diffusion-webui
print("Git safety measure")
!git reset --hard
print("Running launch.py. From there, it's stable-diffusion-webui's code.")
!python launch.py --share --xformers --enable-insecure-extension-access --gradio-queue $host $dark --ckpt $ckpt_arg
```
